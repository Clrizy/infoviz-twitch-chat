```python
import pandas as pd
from datetime import datetime
import re


def get_chat_dataframe(file):
    data = []

    with open(file, 'r', encoding='utf-8') as f:
        lines = f.read().split('\n\n')

        for i, line in enumerate(lines):
            try:
                split_line = line.split('—')
                first_part, second_part = split_line[0], split_line[1:]
                if second_part != []:
                    # how this works if that there is NO "—" character then first_part is the message, second_part is []
                    # if there is NO "—" character, then the time_logged is simply whatever the last time logged was
                    # only update time logged if there is a "—" character (aka second_part is not [])
                    time_logged = first_part.strip()
                    time_logged = datetime.strptime(time_logged, '%Y-%m-%d_%H:%M:%S')

                    username_message = second_part
                    username_message = '—'.join(username_message).strip()

                   
                elif second_part == []:
                    # if there is no second part, we need the username message to be the first part, and
                    # we need to not update the time
                    username_message = first_part

                else:
                    print('wtf')
                
                
                username, channel, message = re.search(
                            ':(.*)\!.*@.*\.tmi\.twitch\.tv PRIVMSG #(.*) :(.*)', username_message
                    ).groups()

                d = {
                    'dt': time_logged,
                    'channel': channel,
                    'username': username,
                    'message': message
                }

                data.append(d)

            except Exception as e:
                print(e)
                pass

    return pd.DataFrame().from_records(data)


df = get_chat_dataframe('loltyler1.log')

df.set_index('dt', inplace=True)

print(df.shape)

df.to_csv('./tyler1_updated.csv')

df.head()
```
